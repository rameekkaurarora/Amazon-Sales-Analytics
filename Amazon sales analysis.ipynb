#### Region with Highest total Sales revenue

```python
Highest_revenue = df.groupby(df['Region'])['Total Revenue'].sum()
Highest_revenue.idxmax()
```

```python
group_data=df.groupby(df['Region'])['Total Revenue'].sum()
sns.set_style('darkgrid')
plt.figure(figsize=(15,5))
sns.lineplot(data=group_data,linestyle='--',linewidth=4,marker='o',markersize=10,markerfacecolor='red',markeredgecolor='black')

plt.xticks(rotation=90)
plt.title('Highest Total Sales Revenue by Region',fontsize=20,color='Blue')
plt.xlabel('Regions',fontsize=20,color='red')
plt.ylabel('Revenue',fontsize=20,color='c')
plt.show()
```

#### Average unit price and unit cost of each item

```python
Unit_price_avg=df.groupby(df['Item Type'])['Unit Price'].mean()
Unit_cost_avg=df.groupby(df['Item Type'])['Unit Cost'].mean()

cost_price_avg=pd.DataFrame({'Average Unit Price':Unit_price_avg,
                              'Average Unit Cost':Unit_cost_avg})

cost_price_avg
                                                
```

#### Country with Highest profit

```python
Total_profit_country=df.groupby(df['Country'])['Total Profit'].sum()
Highest_total_profit_country= Total_profit_country.idxmax()

print("Country with highest total profit is :",Highest_total_profit_country)
```

```python
group_data=df.groupby(df['Country'])['Total Profit'].sum()
sns.set_style('darkgrid')
plt.figure(figsize=(15,5))
sns.barplot(x=group_data.index,y=group_data)

plt.xticks(rotation=90)
plt.title('Country with Highest Total Profit',fontsize=20,color='Blue')
plt.xlabel('Country',fontsize=20,color='red')
plt.ylabel('Profit',fontsize=20,color='c')
plt.show()
```

#### How sales channel affect the order priority distribution

```python
Sales_channel_order_priority_distribution=df.groupby(df['Sales Channel'])['Order Priority'].value_counts()
Sales_channel_order_priority_distribution
```

```python
Sales_channel_order_priority_distribution=df.groupby(df['Sales Channel'])['Order Priority'].value_counts()
Sales_channel_order_priority_distribution=Sales_channel_order_priority_distribution.reset_index(name='Count')

sns.set_style('darkgrid')

plt.figure(figsize=(10,6))
sns.barplot(x='Sales Channel',y='Count',hue='Order Priority',data=Sales_channel_order_priority_distribution)

plt.xlabel('Sales Channel')
plt.ylabel('Count')
plt.title('Sales channel order priority distribution')
plt.show()
```

#### Average order processing time for each sales channel

```python
df['Processing Time']=df['Ship Date']-df['Order Date']

avg_processing_time = df.groupby(df['Sales Channel'])['Processing Time'].mean()
avg_processing_time
```

```python
plt.figure(figsize=(7,6))
sns.barplot(data=df,x=df['Sales Channel'],y=df['Processing Time'].dt.days , width=0.4)

plt.title('Average Processing Time by sales channel')
plt.xlabel('Sales Channel')
plt.yticks(np.arange(0,25,1))
plt.ylabel('Average Processing Time by sales channel')

plt.show()
```

#### Item type having highest and lowest sale

```python
group_item_type=df.groupby(df['Item Type'])['Total Revenue'].sum()

highest_sales_item_type = group_item_type.idxmax()
lowest_sales_item_type = group_item_type.idxmin()
print("Highest sales revenue by item type:",highest_sales_item_type,"\n"'Lowest sales revenue by item type:',lowest_sales_item_type)
```

```python
plt.figure(figsize=(10,5))
sns.scatterplot(x=group_item_type.index,y=group_item_type , s=200)
max_index=group_item_type.idxmax()
plt.scatter(x=max_index,y=group_item_type[max_index],s=200,color='Green',edgecolor='black')

min_index=group_item_type.idxmin()
plt.scatter(x=min_index,y=group_item_type[min_index],s=200,color='Red',edgecolor='black')

plt.xticks(rotation=45)
plt.yticks(rotation=0)

plt.show()
```

#### Corelation between unit price and total profit?

```python
correlation_unit_price_total_profit=df['Unit Price'].corr(df['Total Profit'])
print('Correlation between unit price and total profit',correlation_unit_price_total_profit)
```

```python
plt.figure(figsize=(4,2))
plt.scatter(x=correlation_unit_price_total_profit , y=correlation_unit_price_total_profit,s=200,color='Red')
plt.xticks(np.arange(-1,2,0.5))
plt.yticks(np.arange(-1,2,0.5))
plt.title('Correlation Unit Price Total Profit')
plt.show()
```

#### Seasonal trends or patterns in sales data

```python
month_names= {1: 'Jan',
              2: ' Feb',
              3: 'Mar',
              4 :'Apr',
              5:'May',
              6:'Jun',
              7:'Jul',
              8:'Aug',
              9:'Sep',
              10:'Oct',
              11:'Nov',
              12:'Dec'}

monthly_sales=df.groupby(df['Order Date'].dt.month)['Total Revenue'].sum()
monthly_sales.index=monthly_sales.index.map(month_names)
monthly_sales
```

```python
sns.barplot(x=monthly_sales.index , y=monthly_sales)
plt.title('Month wise total Revenue')
plt.xlabel('Month')
plt.ylabel('Total Revenue')
plt.show()
```

#### Outliers in total cost distribution

```python
q1=df['Total Cost'].quantile(0.25)
q3=df['Total Cost'].quantile(0.75)

iqr=q3-q1
lower_fence=q1-1.5*iqr
upper_fence=q3+1.5*iqr

outliers=df[(df['Total Cost']<lower_fence) | (df['Total Cost']>upper_fence)].reset_index(drop=True)
outliers
```

```python
plt.boxplot(df['Total Cost'])
plt.title('Boxplot of Total Cost')
plt.show()
```
